```python
def calculate_cp_of_two_clusterings(first,second,label): # based on labels O(k^2)
    k=len(first[label].value_counts())
    prob_dict1={}
    prob_dict2={}
    for i in range(0,k):
        prob_dict1[i]=set()
    for i in first.index:
        prob_dict1[int(first.iloc[i][label])].add(i)
    ##print(prob_dict1)
    for i in range(0,k):
        prob_dict2[i]=set()
    for i in second.index:
        prob_dict2[int(second.iloc[i][label])].add(i)
    list_of_prob_dict=[]
    list_of_prob_dict.append(prob_dict1)
    list_of_prob_dict.append(prob_dict2)
    ##print(prob_dict2)
    prob_list=[]
    ##for labels in range(0,k):
        ##print(len(prob_dict1[2].intersection(prob_dict2[labels])))
    for clustering in range(0,2):
        prob_list.append([])
        for label in range(0,k):
            prob_list[clustering].append([])
            for comp_cluster in range(0,2):
                prob_list[clustering][label].append([])
                for lab in range(0,k):
                    temp=len(list_of_prob_dict[clustering][label].intersection(list_of_prob_dict[comp_cluster][lab]))
                    prob_list[clustering][label][comp_cluster].append(temp/len(list_of_prob_dict[comp_cluster][lab]))                 
    return prob_list
```

```python
def log_base_2(num): # Input is of type number
    if num == 0:
        return 0
    elif num > 0 :
        return np.log2(num)
    else :
        print('Entered a negative number')
        return 0
```

```python
def clustering_level_sub_surp(first,second,sec_j,label,lis): # O(k)
    k=len(first[label].value_counts())
    res=0
    first_var_cnt=first[label].value_counts()
    sum_first=first_var_cnt.sum()
    for i in range(0,k):
        res+=(first_var_cnt.iloc[i]/sum_first)*log_base_2(lis[0][i][1][sec_j])
    return -1*res
```

```python
def clustering_level_surp(first,second,label):##k^2
    lis=calculate_cp_of_two_clusterings(first,second,label) ##k^2
    k=len(first[label].value_counts())
    res=0
    second_var_cnt=second[label].value_counts()
    sum_first=second_var_cnt.sum()
    for j in range(0,k):
        res+=(second_var_cnt.iloc[j]/sum_first)*(clustering_level_sub_surp(first,second,j,label,lis))
    return res
```

```python
def clustering_levl_main_surp(list_of_clustering): #O(m^2*k^2)
    clustering_surprisal=[]
    no_of_base_clusterings=len(list_of_clustering)
    for i in range(0,no_of_base_clusterings):
        res=0
        for j in range(0,no_of_base_clusterings):
            if i!=j:
                res+=(clustering_level_surp(list_of_df[i],list_of_df[j],'Cluster'))
        clustering_surprisal.append(res) 
    return clustering_surprisal
```

```python
def cal_for_testing(data,clusters,label):
    res=0
    data_len=len(data)
    clusters_count=data[label].value_counts()
    for i in range(0,clusters):
        res+=(clusters_count[i]/data_len)*log_base_2(clusters_count[i]/data_len)
    return -1*res
```

```python
def cluster_level_surprisal(list_of_df,label):#O(m^2*k^2)
    ##Input take an array of base_clusterings OutPut: Return a list 2D List [clustering][cluster]
    cluster_surprisal_value=[]
    no_of_clusterings=len(list_of_df)
    if(no_of_clusterings==0):
        return cluster_level
    no_of_clusters=len(list_of_df[0][label].value_counts())
    len_of_data=len(list_of_df[0])
    
    for clustering in range(0,no_of_clusterings):
        cluster_surprisal_value.append([0 for i in range(0,no_of_clusters)])
        for int_clustering in range(0,no_of_clusterings):
            if(clustering!=int_clustering):
                prob_cl=calculate_cp_of_two_clusterings(list_of_df[clustering],list_of_df[int_clustering],label)
                for cluster in range(0,no_of_clusters):
                    temp=0
                    for int_cluster in range(0,no_of_clusters):
                        var=log_base_2(prob_cl[0][cluster][1][int_cluster])
                        temp=temp-(list_of_df[clustering][label].value_counts()[int_cluster]/len_of_data)*var
                    cluster_surprisal_value[clustering][cluster]+=temp;              
                      
    return cluster_surprisal_value   
            
```

```python
def weighted_cluster_level_surprisal(cluster_level): # 2D matrix of [clustering][cluster]O(m*k)
    M=len(cluster_level)
    weighted_cluster=[]
    k=len(cluster_level[0])
    for clustering in range(0,M):
        weighted_cluster.append([])
        for cluster in range(0,k):
            weighted_cluster[clustering].append(np.exp((-1/M)*cluster_level[clustering][cluster]))
    return weighted_cluster
```

```python
def cal_association_matrix(cluster_level,list_of_clusterings,label):
    # Input cluster_level weightage ,list_of_df,label
    #Returns the N*N matrix representing the co-association matrix between the data instances in the dataset
    ass_matrix=[]
    no_of_clusterings=len(list_of_clusterings)
    no_of_clusters=len(cluster_level[0])
    no_of_data_inst=len(list_of_clusterings[0])
    list_of_cluster_label=[]
    for i in range(0,no_of_clusterings):
        list_of_cluster_label.append(np.array((list_of_clusterings[i][label])))
    for i in range(0,no_of_data_inst):
        ass_matrix.append([0 for ele in range(0,no_of_data_inst)])
        for j in range(0,no_of_data_inst):
            temp=0
            for clustering in range(0,no_of_clusterings):
                if(list_of_cluster_label[clustering][i]==list_of_cluster_label[clustering][j]):
                    temp+=cluster_level[clustering][int(list_of_cluster_label[clustering][i])]
            ass_matrix[i][j]=temp
    return ass_matrix
```

```python
def find(parent,index):
    if parent[index]==index:
        return index
    parent[index]=find(parent,parent[index])
    return parent[index]
```

```python
def union(a,b,parent,size):
    a=find(parent,a)
    b=find(parent,b)
    if a!=b:
        parent[b]=parent[a]
        size[a]+=size[b]
    return (parent,size)
```

```python
def whac(list_of_df,no_of_clusters):
    cluster_level=cluster_level_surprisal(list_of_df,'Cluster')
    weighted_cluster_value=weighted_cluster_level_surprisal(cluster_level)
    ass_matrix=cal_association_matrix(weighted_cluster_value,list_of_df,'Cluster')
    
    data=list_of_df[0].copy()
    # data is of type dataframe,ass_matrix is of 2D N*N matrix
    no_of_datainstances=len(data)
    p=[0 for i in range(0,no_of_datainstances)]
    c=[0 for i in range(0,no_of_datainstances)]
    active_clusters=[1 for i in range(0,no_of_datainstances)]
    size=[1 for i in range(0,no_of_datainstances)]
    for i in range(0,no_of_datainstances):
        c[i]=i
        p[i]=i

    for a in range(1,no_of_datainstances-no_of_clusters+1):
        s=0
        t=0
        max_value=-100
        for i in range(0,no_of_datainstances):
            for j in range(i+1,no_of_datainstances):
                if active_clusters[i]==1 and active_clusters[j]==1 and max_value<ass_matrix[i][j] and find(c,i)!=find(c,j):
                    max_value=ass_matrix[i][j]
                    s=i
                    t=j
                   # print('parent',find(c,i),find(c,t))
        #print('Max_value',max_value)
        s=find(c,s)
        t=find(c,t)
          
        if(t<s):
            (s,t)=(t,s)
        for u in range(0,no_of_datainstances):
            if c[u] not in (s,t) and active_clusters[u]==1:
                ass_matrix[s][u]=(size[s]*ass_matrix[s][u]+size[t]*ass_matrix[t][u])/(size[s]+size[t])
                ass_matrix[u][s]=ass_matrix[s][u]
        # Union
        
        #print(s,t)
        if s!=t:
            c[t]=c[s]
            size[s]+=size[t]     
        active_clusters[t]=0
        #print(a)
    for i in range(0,no_of_datainstances):
        find(c,i)
    temp_set=set()
    for i in range(0,no_of_datainstances):
        temp_set.add(c[i])
    temp_dict={}
    cluster_count=0
    for i in temp_set:
        temp_dict[i]=cluster_count
        cluster_count+=1
    temp_dict
    temp_data=data
    for i in range(0,no_of_datainstances):
        temp_data.loc[i,'Cluster']=temp_dict[c[i]]
        
    
    return temp_data
```

```python
import math 
    
def calculate_nmi(clustering_1,clustering_2,no_of_clusters,label_1,label_2):
    k=len(clustering_1[label_1].value_counts())
    prob_dict1={}
    prob_dict2={}
    for i in range(0,k):
        prob_dict1[i]=set()
    for i in clustering_1.index:
        prob_dict1[int(clustering_1.iloc[i][label_1])-1].add(i)
    ##print(prob_dict1)
    for i in range(0,k):
        prob_dict2[i]=set()
    for i in clustering_2.index:
        prob_dict2[int(clustering_2.iloc[i][label_2])].add(i)
    
    #### Calulate Numerator
    numerator=0
    N=len(clustering_1)
    for cluster_1 in range(0,k): ## In the first cluster
        for cluster_2 in range(0,k):
            Ni=len(prob_dict1[cluster_1])
            Nj=len(prob_dict2[cluster_2])
            Nij=len(prob_dict1[cluster_1].intersection(prob_dict2[cluster_2]))
            print(Ni,Nj,Nij)
            log_val=log_base_2((Nij*N)/(Ni*Nj))
            numerator+=Nij*log_val
    
    ### First denominator term
    denom_first=0
    for cluster_1 in range(0,k):
        Ni=len(prob_dict1[cluster_1])
        log_val=log_base_2(Ni/N)
        denom_first+=Ni*log_val
        
    ### Second Denominator term
    denom_sec=0
    for cluster_2 in range(0,k):
        Nj=len(prob_dict2[cluster_2])
        log_val=log_base_2(Nj/N)
        denom_sec+=Nj*log_val
    
    ## Calculate Denominator
    print("numerator",numerator,denom_first,denom_sec)
    denominator=math.sqrt(denom_first*denom_sec)
    
    return numerator/denominator;
```
